```
import keras_ocr
import matplotlib.pyplot as plt
```

```
pipeline=keras_ocr.pipeline.Pipeline()
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
images = [keras_ocr.tools.read(url) for url in [
                                          '/content/drive/MyDrive/test_1.jpg',
                                          '/content/drive/MyDrive/test_2.jpg',
                                          '/content/drive/MyDrive/test_5.jpg'
        
    ]
]
```

```
prediction=pipeline.recognize(images)
```

```
prediction
```

```
fig, axs = plt.subplots(nrows=len(images), figsize=(80, 80))

for ax,image,predictions in zip(axs,images,prediction):
  keras_ocr.tools.drawAnnotations(image=image,predictions=predictions,ax=ax)
```

Metric

IOU = Intersection / union
